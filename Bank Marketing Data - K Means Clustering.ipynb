```python
X = pd.get_dummies(df)
```

```python
X
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
scaled_X = scaler.fit_transform(X)
```

```python
from sklearn.cluster import KMeans
```

```python
ssd =[]

for k in range(1,31):
    model = KMeans(n_clusters=k)
    model.fit_predict(scaled_X)
    ssd.append(model.inertia_)
```

```python
ssd
```

```python
plt.figure(figsize=(12,8), dpi=200)
plt.plot(range(1,31), ssd, 'o--')
plt.xlabel('K value')
plt.ylabel('Sum of squared distances')
plt.xticks(np.arange(1, 31, 1.0));
```

```python
plt.figure(figsize=(12,8), dpi=200)
pd.Series(ssd).diff().plot(kind='bar')
```

```python
plt.figure(figsize=(12,8), dpi=200)
pd.Series(ssd).diff().plot(kind='bar')
plt.ylim(-40000, 25000)
```

Points of interest:
Going from K=9, to K=10
Going from K=16, to K=17
Going from K=22, to K=23

16 - emp.var.rate: employment variation rate - quarterly indicator (numeric)<br>
19 - euribor3m: euribor 3 month rate - daily indicator (numeric)

```python
sns.scatterplot(data=df, x='emp.var.rate', y='euribor3m')
```

```python

def run_model(k, scaled_X):
    model = KMeans(n_clusters=k)
    cluster_labels = model.fit_predict(scaled_X)
    X['Cluster'] = cluster_labels
    X.corr()['Cluster'][:-1].sort_values()
    print(X.corr()['Cluster'][:-1].sort_values())

    plt.figure(figsize=(12,8), dpi=200)
    sns.barplot(y=X.corr()['Cluster'][:-1].sort_values(), x=X.corr()['Cluster'][:-1].sort_values().index)
    plt.xticks(rotation=90);
```

```python
run_model(16, scaled_X)
```

```python
sns.countplot(data=df, x='poutcome', hue='contact')
```

15 - poutcome: outcome of the previous marketing campaign (categorical: 'failure','nonexistent','success')<br>
8 - contact: contact communication type (categorical: 'cellular','telephone')<br>
With K=16, previous outcome and type of contact seems to suggest there may be clusters present.

```python
run_model(22, scaled_X)
```

month: last contact month of year (categorical: 'jan', 'feb', 'mar', ..., 'nov', 'dec')
